```python
import cv2
import numpy as np
import scipy as sp
from scipy import signal
from matplotlib import pyplot as plt
```

```python
# Loading road image and making a grayscale version
img_road_bgr = cv2.imread("road.jpg")
img_road_rgb = cv2.cvtColor(img_road_bgr, cv2.COLOR_BGR2RGB)
img_road_gray = cv2.cvtColor(img_road_rgb, cv2.COLOR_RGB2GRAY)
```

```python
# Displaying the RGB image and the grayscale version
plt.figure(figsize=(20,7))

plt.subplot(121)
plt.imshow(img_road_rgb)
plt.title('Input image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(img_road_gray, cmap="gray")
plt.title('Grayscale image')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Applying gaussian blur to the image
# gaussian2_xy and gaussian2d(2_n) are from Computer Vision - Image Filtering python notebook
def gaussian2_xy(mean, cov, xy):
    invcov = np.linalg.inv(cov)
    results = np.ones([xy.shape[0], xy.shape[1]])
    for x in range(0, xy.shape[0]):
        for y in range(0, xy.shape[1]):
            v = xy[x,y,:].reshape(2,1) - mean
            results[x,y] = np.dot(np.dot(np.transpose(v), invcov), v)
    results = np.exp( - results / 2 )
    return results 

def gaussian2d(mu, cov, n):
    s = int(n/2)
    x = np.linspace(-s,s,n)
    y = np.linspace(-s,s,n)
    xc, yc = np.meshgrid(x, y)
    xy = np.zeros([n, n, 2])
    xy[:,:,0] = xc
    xy[:,:,1] = yc

    return gaussian2_xy(mu, cov, xy), xc, yc

n = 3
mean =  np.array([0, 0])
mean = mean.reshape(2,1)
cov = np.array([[20,0],[0,20]])

g2d_kernel, xc, yc = gaussian2d(mean, cov, n)
g2d_kernel_normalized = g2d_kernel / np.sum(g2d_kernel)
img_road_blurred = signal.convolve2d(img_road_gray, g2d_kernel_normalized, mode='same', boundary='fill')
```

```python
# Displaying the grayscale image and the blurred version
plt.figure(figsize=(20,7))

plt.subplot(121)
plt.imshow(img_road_gray, cmap="gray")
plt.title('Input image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(img_road_blurred, cmap="gray")
plt.title('Grayscale image')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
# Using canny edge detector to get edges of image
img_road_blurred_u8 = np.uint8(img_road_blurred)
img_edges = cv2.Canny(img_road_blurred_u8, 150, 150)
```

```python
# Displaying the blurred image and the canny edge image
plt.figure(figsize=(20,7))

plt.subplot(121)
plt.imshow(img_road_blurred, cmap="gray")
plt.title('Input image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(img_edges, cmap="gray")
plt.title('Grayscale image')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
tri_poly = np.array([
                        [
                            (img_edges.shape[1]//8, img_edges.shape[0]),
                            (img_edges.shape[1]//8 * 7, img_edges.shape[0]), 
                            (img_edges.shape[1]//2, img_edges.shape[0]//8 * 5)
                        ]
                    ])
mask = np.zeros((img_edges.shape[0], img_edges.shape[1]))
mask = np.uint8(mask)
cv2.fillPoly(mask, tri_poly, 255)
img_interest = cv2.bitwise_and(img_edges, mask)
```

```python
# Displaying the canny edge image and region of interest
plt.figure(figsize=(20,7))

plt.subplot(121)
plt.imshow(img_edges, cmap="gray")
plt.title('Canny image')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(img_interest, cmap="gray")
plt.title('Region of interest')
plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import numpy as np
import cv2 as cv
import math

laneVideo = cv.VideoCapture("LaneDetectionTestVideo01.mp4")
ok, frame = laneVideo.read()

width = frame.shape[1]
height = frame.shape[0]

width_l = width // 7
width_r = width - width // 7

height_b = height
height_t = height // 11 * 8

# Triangle polygon
tri_poly = np.array([
                        [
                            (width_l, height_b),
                            (width_r, height_b), 
                            ((width_r + width_l) // 2, height_t)
                        ]
                    ])

mask = np.zeros((frame.shape[0], frame.shape[1]))
mask = np.uint8(mask)
cv.fillPoly(mask, tri_poly, 255)

# White colour range (RGB)
lower_w = np.array([100,100,200], dtype=np.uint8)
upper_w = np.array([255,255,255], dtype=np.uint8)

# Yellow colour range (RGB)
lower_y = np.array([140, 130, 70], dtype=np.uint8)
upper_y = np.array([250, 240, 120], dtype=np.uint8)

# read the video
while (ok):
    # Filtering the frame using a white and yellow mask
    frame_rgb = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
    white_mask = cv.inRange(frame_rgb, lower_w, upper_w)
    yellow_mask = cv.inRange(frame_rgb, lower_y, upper_y)
    both_masks = cv.bitwise_or(white_mask, yellow_mask)
    frame_masked = cv.bitwise_and(frame, frame, mask=both_masks)
    frame_gray = cv.cvtColor(frame_masked, cv.COLOR_RGB2GRAY)

    frame_blur = cv.GaussianBlur(frame_gray, (5, 5), 0)
    frame_blur_u8 = np.uint8(frame_blur)
    frame_edges = cv.Canny(frame_blur_u8, 100, 300)  
    frame_lane = cv.bitwise_and(frame_edges, mask)

    # https://docs.opencv.org/2.4/modules/imgproc/doc/feature_detection.html?highlight=houghlinesp#houghlinesp
    # cv2.HoughLinesP(image, rho, theta, threshold[, lines[, minLineLength[, maxLineGap]]])  lines
    lines = cv.HoughLinesP(frame_lane, 2, np.pi / 180, 50, lines = np.array([]), minLineLength = 25, maxLineGap = 40)
    
    if lines is not None:
        for line in lines:
            # Getting coordinates from line to draw it
            x1, y1, x2, y2 = line[0]
            angle = math.atan2(y2 - y1, x2 - x1) * 180.0 / np.pi;
            
            # Ignoring lines that are close to horizontal
            if(angle < -15 or angle > 15):
                # Drawing a line in frame_gray based on arguments
                cv.line(frame, (x1, y1), (x2, y2), (255, 0, 0), 3)
        
    cv.imshow("Video", frame_lane)

    if (cv.waitKey(1) & 0xFF == ord('q')):
        break

    ok, frame = laneVideo.read() # return value of ok, and frame
        
# The following frees up resources and closes all windows
laneVideo.release()
cv.destroyAllWindows()
```
